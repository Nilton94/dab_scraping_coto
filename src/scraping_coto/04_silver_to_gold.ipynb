# Imports

```python
import sys 
dbutils.widgets.text(name='source_path', defaultValue='.')
source_path = dbutils.widgets.get("source_path")
sys.path.append(source_path)
from scraping_coto.etl.gold.silver_to_gold import load_gold_items_table
from modules.utils.logger import get_logger

logger = get_logger("silver_to_gold")
```

# Configs

```python
logger.info("Starting Silver to Gold ETL process")

dbutils.widgets.text(name='target', defaultValue='dev')
bundle_target = dbutils.widgets.get("target")

dbutils.widgets.text('source_catalog_name', 'default_catalog')
source_catalog_name = dbutils.widgets.get('source_catalog_name') + '_' + bundle_target if bundle_target != 'prod' else dbutils.widgets.get('source_catalog_name')

dbutils.widgets.text('source_schema_name', 'default_schema')
source_schema_name = dbutils.widgets.get('source_schema_name')

dbutils.widgets.text('source_table_name', 'default_table')
source_table_name = dbutils.widgets.get('source_table_name')

dbutils.widgets.text('target_catalog_name', '')
target_catalog_name = dbutils.widgets.get('target_catalog_name')  + '_' + bundle_target if bundle_target != 'prod' else dbutils.widgets.get('target_catalog_name')

dbutils.widgets.text('target_schema_name', '')
target_schema_name = dbutils.widgets.get('target_schema_name')

dbutils.widgets.text('target_table_name', '')
target_table_name = dbutils.widgets.get('target_table_name')

dbutils.widgets.text('partition_by', '')
partition_by = dbutils.widgets.get('partition_by').split(',') if dbutils.widgets.get('partition_by') else None

dbutils.widgets.text('mode', 'append')
mode = dbutils.widgets.get('mode')

dbutils.widgets.text('date_column', 'dt')
date_column = dbutils.widgets.get('date_column')

logger.info(f"Configgurations:")
logger.info(f"  - Bundle target: {bundle_target}")
logger.info(f"  - Source: {source_catalog_name}.{source_schema_name}.{source_table_name}")
logger.info(f"  - Target: {target_catalog_name}.{target_schema_name}.{target_table_name}")
logger.info(f"  - Partition by: {partition_by}")
logger.info(f"  - Mode: {mode}")
logger.info(f"  - Date column: {date_column}")
```

# Load

```python
load_gold_items_table(
    source_catalog_name=source_catalog_name,
    source_schema_name=source_schema_name,
    source_table_name=source_table_name,
    target_catalog_name=target_catalog_name,
    target_schema_name=target_schema_name,
    target_table_name=target_table_name,
    partition_by=partition_by,
    mode=mode,
    date_column=date_column
)
```
