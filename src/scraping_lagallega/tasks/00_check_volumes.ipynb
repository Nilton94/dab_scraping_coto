# Code

## Check Catalogs

```python
target = 'lagallega' if bundle_target == 'prod' else f'lagallega_{bundle_target}'
```

```python
spark.sql(f'CREATE CATALOG IF NOT EXISTS {target} COMMENT "Catálogo para os dados extraídos do site La Gallega na camada de {bundle_target}";')
spark.sql(f'CREATE SCHEMA IF NOT EXISTS {target}.bronze COMMENT "Camada Bronze para os Dados do La Gallega";')
spark.sql(f'CREATE SCHEMA IF NOT EXISTS {target}.silver COMMENT "Camada Silver para os Dados do La Gallega";')
spark.sql(f'CREATE SCHEMA IF NOT EXISTS {target}.gold COMMENT "Camada Gold para os Dados do La Gallega";')
spark.sql(f'CREATE VOLUME IF NOT EXISTS {target}.bronze.staging COMMENT "Staging para dados macro de departamentos e suas categorias/subcategorias";')
spark.sql(f'CREATE VOLUME IF NOT EXISTS {target}.silver.checkpoints COMMENT "Volume para checkpoints dos dados das tabelas silver";')
```
